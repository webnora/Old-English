```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_calgary = pd.read_csv('frequency_calgary.csv',index_col='Unnamed: 0')
df_calgary = df_calgary.drop(['all','unique','unique_stop'],1)
df_calgary
```

```python
df_ycoe = pd.read_csv('frequency_ycoe.csv', index_col='Unnamed: 0')
df_ycoe = df_ycoe.drop(['all','unique','unique_stop'],1)
df_ycoe
```

```python
result_table = pd.concat([df_ycoe, df_calgary], axis=1, sort=False)
result_table = result_table.rename(columns={'all_words_stop': 'tokens frequency in prose (ycoe)', 'all_stop': 'tokens frequency in poetry (calgary)'}).fillna(0).astype(int).sort_values('tokens frequency in prose (ycoe)', ascending=False)
result_table
```

```python
result_table['percentage ycoe'] = (result_table['tokens frequency in prose (ycoe)']/result_table['tokens frequency in prose (ycoe)'].sum()*100).round(2)
result_table['percentage calgary'] = (result_table['tokens frequency in poetry (calgary)']/result_table['tokens frequency in poetry (calgary)'].sum()*100).round(2)
# result_table = result_table.astype(int)
result_table
```

```python
result_table[['percentage ycoe','percentage calgary']].plot(kind='bar', figsize=(15, 10))
# plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
```
