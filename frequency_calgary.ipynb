```python
from nltk.corpus import PlaintextCorpusReader
corpus_root = '../Calgary_corpus'
wordlists = PlaintextCorpusReader(corpus_root, '.*')
# wordlists.fileids()
```

```python
words_calgary = wordlists.words()
words_calgary[0:10]
len(words_calgary)
```

```python
from cltk.stop.old_english.stops import STOPS_LIST
calgary_stop = [w for w in words_calgary if not w in STOPS_LIST]
len(calgary_stop)
```

```python
from collections import defaultdict
dic_words = defaultdict(list)
for word in words_calgary:
    initial = word[0] 
    if initial not in ['#','.',',',';','-','—',':','"','?',')','(','!', "'"]:
        dic_words[initial].append(word)
dic_words.keys()
```

```python
# type(dic_words)
str1 = ' '.join(str(e) for e in words_calgary)
str1 [0:10]
```

```python
import re
re.findall(r"\s(v\w*)", str1)
```

```python
frequency = {}
for key in dic_words:
    frequency[key] = {'all': len(dic_words[key]), 'unique': len(set(dic_words[key]))}
table_poesy = pd.DataFrame.from_dict(frequency, orient='index').sort_values('unique', ascending=False)
# table_poesy
```

```python
from collections import defaultdict
dic_words_stop = defaultdict(list)
for word in calgary_stop:
    initial = word[0] 
    if initial not in ['#','.',',',';','-','—',':','"','?',')','(','!', "'"]:
        dic_words_stop[initial].append(word)
dic_words_stop.keys()
```

```python
frequency_stop = {}
for key in dic_words_stop:
    frequency_stop[key] = {'all_stop': len(dic_words_stop[key]), 'unique_stop': len(set(dic_words_stop[key]))}
table_poesy_stop = pd.DataFrame.from_dict(frequency_stop, orient='index').sort_values('unique_stop', ascending=False)
# table_poesy
```

```python
# table_poesy_stop
```

```python
result = table_poesy.join(table_poesy_stop)
result
```
